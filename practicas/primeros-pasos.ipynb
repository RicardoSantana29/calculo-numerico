<h3>Marco Teórico</h3>
Las gráficas se emplean para representar funciones, ecuaciones, modelos matemáticos o
situaciones que expresadas de manera gráfica ayudan a su interpretación.

Recordemos que en matemática $y = f(x)$ nos indica que se trata de una función cuyo
dominio se asocia a la variable $x$ y el rango o imágenes a la variable $y$. En principio estas
imágenes se obtienen al realizar los cálculos especificados por la regla de asignación
expresada en $f(x)$.

Ejemplos de funciones:

1. $y = f(x) = x$ tal que $x\in \mathbb{Z}$, una función discreta.
2. $y = f(x) = x$ tal que $x\in \mathbb{R}$, una función continua.

<h3>Práctica</h3>
<strong>1)</strong> Se procede a importar el módulo <strong>pylab</strong> y todas sus funciones.

```python
import pylab as lab
x=lab.arange(10)
dibujo_1=lab.plot(x,'-o')
```

<strong>Análisis</strong> Se representa una función de variable discreta cuyos puntos se ha unido con una recta, La función se ha definido de manera implicita con el comando plot.

Acá el dominio de la función es $\{x\in \mathbb{Z}:x = 1, ... , 10\}$ y el rango es $\{y\in \mathbb{Z}:x = 1, ... , 10\}$, es claro que la regla de asignación es $y = x$.

<strong>2)</strong> Se utiliza el código anterior y se importa la módulo (librería) <strong>numpy</strong> que permite realizar

```python
import pylab as lab
import numpy as np
x=lab.arange(10)
dibujo_1=lab.plot(2*x,'bx-')
```

<strong>Análisis</strong> Al igual que en el numeral, <strong>1)</strong>, estamos dibujando una función discreta en la que se ha definido de manera implicita a travez del comando <strong>plot</strong> la función

<strong>3)</strong> La instrucción <strong>clf()</strong> se utiliza para borrar de memoria todo lo relacionado con las figuras anteriores, observe que como ya se importaron y ejecutaron las librerías que nos permiten realizar gráficas y operar sobre variables, en este caso, no son llamadas.
```python
lab.clf()
x2 = x**2
x3 = x**3
lab.plot(x,x,'b.',x,x2,'rd',x,x3,'g^')
lab.xlim(-1,11)
lab.ylim(-50,850)
```
