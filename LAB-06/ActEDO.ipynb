#Parte 1

```python
def f(t,x):
  return -t*x+(t**3)*(x**3)
import numpy as np
import matplotlib.pyplot as plt
```

#Parte 2

```python
#Metodo de Euler
def Euler(fun, t0, x0, h, m):
    # fun: nombre de la función f(t,x)
    # Condición inicial x(t0) = x0;
    # Tamaño de paso: h > 0
    # Número de puntos: m (entero positivo)
    t = t0 + h*np.arange(m + 1)
    x = np.zeros(m + 1)
    x[0] = x0
    for i in range(m):
        x[i + 1] = x[i] + h * fun(t[i], x[i])
    return x, t

E,T=Euler(f,0,0.5,0.3,10)
print(E,T)
```

```python
#Runge kutta de 4to orden
def RungeKutta4(fun, t0, x0, h, m):
    t = np.arange(t0, t0 + h*(m+1), h)
    x = np.zeros(m+1)
    x[0] = x0

    for i in range(m):
        k1 = h*fun(t[i], x[i])
        k2 = h*fun(t[i]+h/2, x[i]+k1/2)
        k3 = h*fun(t[i]+h/2, x[i]+k2/2)
        k4 = h*fun(t[i]+h, x[i]+k3)
        x[i+1] = x[i] + (k1 + 2*k2 + 2*k3 + k4)/6

    return x, t
R,t=RungeKutta4(f, 0, 0.5, 0.3, 10)
print(R,t)
```

```python
"""
Dividir intervalo en 100 partes iguales, presentar los resultados
apropiadamente en una tabla y graficar
"""
from tabulate import tabulate
r,t=RungeKutta4(f, 0, 0.5, 3/100, 100)
table0=[]
e,t=Euler(f, 0, 0.5, 3/100, 100)
for i in range(101):
  table0.append([r[i],t[i],e[i]])
print("Tabla de resultados")
print(tabulate(table0, headers = ['valor solucion runge kutta','iteracion','valor solucion Euler' ]))

plt.grid()
plt.plot(t,r,'b-')
plt.title('Grafico para Runge-kutta con respecto al tiempo')
plt.ylabel('valor solucion')
plt.xlabel('tiempo')
plt.show()

plt.grid()
plt.plot(t,e,'r-')
plt.title('Grafico para Euler con respecto al tiempo')
plt.ylabel('valor solucion')
plt.xlabel('tiempo')
plt.show()
```

#Parte 3
EDO $x′′(t) + 4x′(t) + 5x(t) = 0$

Se aplica el cambio de variable $X_1=X$, $X_2=X'$
por tanto el sistema quedaria de la siguiente forma

$\left\{\begin{matrix}
 X'_{1}=X_2\\
  X'_2=- 4X_2 - 5X_1\\
 \end{matrix}\right.$

```python
def F(t,x):
    return np.array([x[1],-4*x[1]-5*x[0]])
def runge_kutta_edo(fun, t0, x0, h, m):
    # Definir el vector  t
    t = t0 + h * np.arange(m+1)

    # Obtener la dimensión del sistema
    n = len(x0)

    # Inicializar la matriz de soluciones
    x = np.zeros((m+1, n))
    x[0, :] = x0

    # Aplicar el método de Runge-Kutta
    for k in range(m):
        k1 = h * fun(t[k], x[k, :])
        k2 = h * fun(t[k] + h/2, x[k, :] + k1/2)
        k3 = h * fun(t[k] + h/2, x[k, :] + k2/2)
        k4 = h * fun(t[k] + h, x[k, :] + k3)

        x[k+1, :] = x[k, :] + (k1 + 2*k2 + 2*k3 + k4) / 6

    return x, t
to=0
m=50
xo=[3,-5]
X,T=runge_kutta_edo(F,to , xo,0.1, m)
print(X)
table1=[]
for i in range(51):
  table1.append([X[i][0],T[i],X[i][1]])

print("Tabla de resultados")
print(tabulate(table1, headers = ['valor de solucion 1','iteracion','valor solucion 2' ]))
plt.grid()
R=np.zeros((51))
R2=np.zeros((51))
for i in range(51):
  R[i]=X[i][0]
  R2[i]=X[i][1]

plt.plot(T,R,'b-')
plt.plot(T,R2,'r-')
plt.legend(('Para X1', 'Para  X2'))
plt.title('Grafico para Sistema de EDOs')
plt.ylabel('valor solucion')
plt.xlabel('tiempo')
plt.show()

```
